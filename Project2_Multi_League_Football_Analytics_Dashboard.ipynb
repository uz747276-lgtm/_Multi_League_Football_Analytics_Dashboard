```python
# Part 1 : Data Collection Fetch standings and scorers for all 3 leagues.
#  Same pattern as Project 1 but 3x.

import requests
import pandas as pd
import numpy as np

# Standings
url_pl = "https://api.football-data.org/v4/competitions/PL/standings"   # Premier League
url_la="https://api.football-data.org/v4/competitions/PD/standings"   # La Liga
url_bl ="https://api.football-data.org/v4/competitions/BL1/standings"  # Bundesliga

# Scorers
score_pl = "https://api.football-data.org/v4/competitions/PL/scorers"     # Premier League
score_la= "https://api.football-data.org/v4/competitions/PD/scorers"     # La Liga
score_bl ="https://api.football-data.org/v4/competitions/BL1/scorers"    # Bundesliga

# HEADERS (all same)
headers = {
    "X-Auth-Token": "cd007f45d0314e788bc11a08847fcc9c"
}

# PARAMS (all same)
params = {
    "season": "2025"
}

#PL standings
res_pl_stand = requests.get(url_pl,headers=headers,params=params)
data_stand_pl = res_pl_stand.json()
loc_pl = data_stand_pl["standings"][0]["table"]
df_stand_pl = pd.json_normalize(loc_pl)
df_stand_pl['league'] = 'PL'
# print(df_stand_pl)

#Laliga standings
res_la_stand = requests.get(url_la,headers=headers,params=params)
data_stand_la = res_la_stand.json()
loc_la = data_stand_la["standings"][0]["table"]
df_stand_la = pd.json_normalize(loc_la)
df_stand_la['league'] = 'La Liga'
#print(df_stand_la)

#Bundesliga Standings
res_bl_stand = requests.get(url_bl,headers=headers,params=params)
data_stand_bl = res_bl_stand.json()
loc_bl = data_stand_bl["standings"][0]["table"]
df_stand_bl = pd.json_normalize(loc_bl)
df_stand_bl['league'] = 'BL1'
#print(df_stand_bl)

#PL scorer
res_pl_score = requests.get(score_pl,headers=headers,params=params)
data_score_pl = res_pl_score.json()
loc_PL = data_score_pl["scorers"]
df_score_pl =pd.json_normalize(loc_PL)
df_score_pl['league'] = 'PL'
#print(df_score_pl)


#Laliga scorer
res_la_score = requests.get(score_la,headers=headers,params=params)
data_score_la = res_la_score.json()
loc_LA = data_score_la["scorers"]
df_score_la =pd.json_normalize(loc_LA)
df_score_la['league'] = 'La Liga'
#print(df_score_la)


#BundesLiga scorer
res_bl_score = requests.get(score_bl,headers=headers,params=params)
data_score_bl = res_bl_score.json()
loc_BL = data_score_bl["scorers"]
df_score_bl =pd.json_normalize(loc_BL)
df_score_bl['league'] = 'BL1'
#print(df_score_bl)


```

```python
#Part 3 — Combining Use pd.concat() to stack all 3 standings into one

# Combine all standings into 1 DF
df_standings_all = pd.concat([df_stand_pl, df_stand_la, df_stand_bl], ignore_index=True)

# Combine all scorers into 1 DF
df_scorers_all = pd.concat([df_score_pl, df_score_la, df_score_bl], ignore_index=True)

# print(df_standings_all)
# print(df_scorers_all)
```

```python
# Part 4 — Analysis

# Top 5 scorers across all 3 leagues combined




#PL top scorer
df_score_pl_clean = df_score_pl[["player.name","team.name","goals","playedMatches"]].copy()
df_score_pl_clean.columns = ["Name","Team","Goals","Matches"]
#print(df_score_pl_clean.head(5))


#laliga Top scorer
df_score_la_clean = df_score_la[["player.name","team.name","goals","playedMatches"]].copy()
df_score_la_clean.columns = ["Name","Team","Goals","Matches"]
#print(df_score_la_clean.head(5))

#BL top scorer

df_score_bl_clean = df_score_bl[["player.name","team.name","goals","playedMatches"]].copy()
df_score_bl_clean.columns = ["Name","Team","Goals","Matches"]
print(df_score_bl_clean.head(5))

print("="*40)
print("PREMIER LEAGUE 2025/26 DASHBOARD")
print("="*40)
print(df_score_pl_clean.head(5))

print("="*40)
print("Laliga 2025/26 DASHBOARD")
print("="*40)
print(df_score_la_clean.head(5))

print("="*40)
print("BundesLiga 2025/26 DASHBOARD")
print("="*40)
print(df_score_bl_clean.head(5))
```

```python
# Which league has the highest average goals per team

#PL Avg goal

# print(df_stand_pl)
totalgoals_pl = (df_stand_pl["goalsFor"]).sum()
total_matches_pl = df_stand_pl.shape[0]
result_pl = totalgoals_pl / total_matches_pl
# print(result_pl)


#LaLiga Avg goal
totalgoals_la = (df_stand_la["goalsFor"]).sum()
total_matches_la = df_stand_la.shape[0]
result_la = totalgoals_la / total_matches_la
# print(result_la)



#BundesLiga Avg goal
totalgoals_bl = (df_stand_bl["goalsFor"]).sum()
total_matches_bl = df_stand_bl.shape[0]
result_bl = totalgoals_bl / total_matches_bl
# print(result_bl)

avg ={
    "PL" : result_pl,
    "LA" : result_la,
    "BL" : result_bl
}

best_league = max(avg,key = avg.get)
best_avg = avg[best_league]

print(f"League with highest avg goals/team: {best_league} : {best_avg}")
```

```python
# Top team from each league by points
#print(df_stand_pl)

#Pl top Team

team_name_pl = df_stand_pl["team.name"][0]
print(f"PL Top Team : {team_name_pl} - {team_points_pl}")

#Laliga top team
team_points_la = df_stand_la["points"][0]
team_name_la = df_stand_la["team.name"][0]
print(f"Laliga Top Team : {team_name_la} - {team_points_la}")

#Bl top team
team_points_bl = df_stand_bl["points"][0]
team_name_bl = df_stand_bl["team.name"][0]
print(f"BL1 Top Team : {team_name_bl} - {team_points_bl}")
```

```python
# Which league has the most competitive gap (difference between 1st and 5th place points)

#PL Diff
team_num1_pl = df_stand_pl["points"][0]
team_num5_pl = df_stand_pl["points"][4]
diff_pl = team_num1_pl-team_num5_pl
#Laliga Diff
team_num1_la = df_stand_la["points"][0]
team_num5_la = df_stand_la["points"][4]
diff_la = team_num1_la-team_num5_la

#Bundesliga diff
team_num1_bl = df_stand_bl["points"][0]
team_num5_bl = df_stand_bl["points"][4]
diff_bl = team_num1_bl-team_num5_bl

diff = {
    "PL" : diff_pl,
    "LA" : diff_la,
    "BL" : diff_bl
}

best_diff_league = min(diff,key=diff.get)
best_diff  = diff[best_diff_league]

print(f"PL : {diff_pl} points gap.")
print(f"Laliga : {diff_la} points gap.")
print(f"Bundesliga : {diff_bl} points gap.")
print(f"Most competitive League : {best_diff_league}")
```

```python

# Part 5: Output - Dashboard print + CSV export

# Top 5 scorers across all leagues
top5_scorers = df_scorers_all.nlargest(5, 'goals')[['player.name', 'goals', 'league']]

# Dashboard print
print("="*60)
print("MULTI-LEAGUE FOOTBALL ANALYTICS REPORT")
print("="*60)

print("\nTOP 5 SCORERS (All Leagues)")
for idx, (i, row) in enumerate(top5_scorers.iterrows(), 1):
    print(f"{idx}. {row['player.name']} ({row['league']}) - {row['goals']} goals")

print("\nAVERAGE GOALS PER TEAM")
print(f"PL: {result_pl:.2f}")
print(f"La Liga: {result_la:.2f}")
print(f"BL1: {result_bl:.2f}")

print("\nTOP TEAM BY POINTS")
print(f"PL: {team_name_pl} - {team_points_pl}")
print(f"La Liga: {team_name_la} - {team_points_la}")
print(f"BL1: {team_name_bl} - {team_points_bl}")

print("\nCOMPETITIVE GAP (1st vs 5th place)")
print(f"PL: {diff_pl} points")
print(f"La Liga: {diff_la} points")
print(f"BL1: {diff_bl} points")
print(f"\nMost Competitive: {best_diff_league}")

print("="*60)

# Save to CSV
df_standings_all.to_csv('standings_all_leagues.csv', index=False)
df_scorers_all.to_csv('scorers_all_leagues.csv', index=False)

# Analysis summary CSV
analysis_data = {
    'Metric': ['Top Scorer', 'Avg Goals/Team', 'Top Team', 'Competitive Gap'],
    'PL': [top5_scorers.iloc[0]['player.name'], f"{result_pl:.2f}", f"{team_name_pl} ({team_points_pl})", diff_pl],
    'La Liga': ['-', f"{result_la:.2f}", f"{team_name_la} ({team_points_la})", diff_la],
    'BL1': ['-', f"{result_bl:.2f}", f"{team_name_bl} ({team_points_bl})", diff_bl]
}

df_analysis = pd.DataFrame(analysis_data)
df_analysis.to_csv('analysis_summary.csv', index=False)

print("\n✓ CSV files saved:")
print("  - standings_all_leagues.csv")
print("  - scorers_all_leagues.csv")
print("  - analysis_summary.csv")
```
